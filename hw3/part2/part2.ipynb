```python
import moviepy.editor as mp
import torch
from transformers import pipeline
from docx import Document

# Define device (CUDA if available)
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the Whisper model for speech recognition
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-small",
    chunk_length_s=30,
    device=device,
)

# Function to extract audio from video
def extract_audio_from_video(video_file, audio_output_file="temp_audio.wav"):
    video_clip = mp.VideoFileClip(video_file)
    video_clip.audio.write_audiofile(audio_output_file)
    return audio_output_file

# Function to generate captions from audio
def generate_captions(audio_file):
    transcription = pipe(audio_file)["text"]
    return transcription

# Function to save captions to a DOCX file
def save_captions_to_doc(captions, output_doc_file="captions.docx"):
    doc = Document()
    doc.add_heading('Video Captions', 0)
    doc.add_paragraph(captions)
    doc.save(output_doc_file)
    print(f"Captions saved to {output_doc_file}")

# Main process
def video_to_captions(video_file, output_doc_file="captions.docx"):
    try:
        # Extract audio from the video
        audio_file = extract_audio_from_video(video_file)
        # Generate captions from the extracted audio
        captions = generate_captions(audio_file)
        # Save the captions to a DOCX file
        save_captions_to_doc(captions, output_doc_file)
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
video_file = "Application_of_Deep_Learning_Course.mp4"  # Replace with your video file path
output_doc_file = "captions.docx"  # Output document file path

video_to_captions(video_file, output_doc_file)
```
